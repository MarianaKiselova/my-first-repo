Statistical Analysis:
We perform a t-test to check whether the price differences between customer segments are statistically significant.

```python
from scipy.stats import ttest_ind

# Split data into two groups: Mainstream vs Premium & Budget
mainstream_prices = data[data["PREMIUM_CUSTOMER"] == "Mainstream"]["PRICE_PER_PACK"]
premium_budget_prices = data[data["PREMIUM_CUSTOMER"] != "Mainstream"]["PRICE_PER_PACK"]

# Perform t-test
t_stat, p_value = ttest_ind(mainstream_prices, premium_budget_prices, equal_var=False)
print(f"T-test p-value: {p_value}")
```

Mainstream customers have a significantly different (likely higher) price per pack for chips compared to customers in the Premium and Budget segments.

The difference in average price per pack is very likely to be real and not due to random chance.

Insights:Mainstream midage and young singles and couples are more willing to pay more per

This is also supported by there being fewer premium midage and young singles and

As the difference in average price per unit isn't large, we can check if this
difference is statistically different

```python
from scipy import stats

# Filter the data for relevant customer segments
mainstream = data[data['PREMIUM_CUSTOMER'] == 'Mainstream']
premium = data[data['PREMIUM_CUSTOMER'] == 'Premium']
budget = data[data['PREMIUM_CUSTOMER'] == 'Budget']
midage_young_singles_couples = data[data['LIFESTAGE'].isin(['MIDAGE SINGLES/COUPLES', 'YOUNG SINGLES/COUPLES'])]

# Perform t-test for Mainstream vs Premium customer segments
t_stat_mainstream_vs_premium, p_value_mainstream_vs_premium = stats.ttest_ind(mainstream['price_per_unit'], premium['price_per_unit'])

# Perform t-test for Budget vs Midage young singles and couples
t_stat_budget_vs_midage, p_value_budget_vs_midage = stats.ttest_ind(budget['price_per_unit'], midage_young_singles_couples['price_per_unit'])

# Print the results
print(f"T-test results for Mainstream vs Premium:")
print(f"T-statistic: {t_stat_mainstream_vs_premium}, P-value: {p_value_mainstream_vs_premium}")

print(f"\nT-test results for Budget vs Midage young singles and couples:")
print(f"T-statistic: {t_stat_budget_vs_midage}, P-value: {p_value_budget_vs_midage}")
```

Both comparisons (Mainstream vs Premium, Budget vs Midage young singles and couples) show significant differences in average price per unit.

The results suggest that Mainstream customers and Midage young singles and couples are willing to pay significantly higher prices for chips compared to the Premium and Budget segments.

```python
import itertools

transaction_column = 'TXN_ID'  
brand_column = 'BRAND'  

# Group the data by transaction ID and get the list of brands for each transaction
transaction_groups = data.groupby(transaction_column)[brand_column].apply(list)

# Create a list to store brand pairs
brand_pairs = []

# Loop through each transaction group to create pairs for multi-brand transactions
for brands_in_transaction in transaction_groups:
    if len(brands_in_transaction) > 1:  # Only consider transactions with more than one brand
        pairs = list(itertools.combinations(brands_in_transaction, 2))  # Create all combinations of brand pairs
        brand_pairs.extend(pairs)

# Convert the brand pairs into a DataFrame
pair_df = pd.DataFrame(brand_pairs, columns=['Brand1', 'Brand2'])

# Calculate the frequency (count) of each brand pair
pair_counts = pair_df.groupby(['Brand1', 'Brand2']).size().reset_index(name='count')

# Calculate the support for each pair (count of pairs / total number of transactions)
pair_counts['support'] = pair_counts['count'] / len(data)

# Display the most frequent brand pairs with support
print(pair_counts.sort_values(by='support', ascending=False).head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Filter out target segment data
target_segment_data = data[
    (data["LIFESTAGE"] == "YOUNG SINGLES/COUPLES") & 
    (data["PREMIUM_CUSTOMER"] == "Mainstream")]

# Filter out the rest of the population (excluding target segment)
rest_of_population_data = data[
    (data["LIFESTAGE"] != "YOUNG SINGLES/COUPLES") & 
    (data["PREMIUM_CUSTOMER"] != "Mainstream")]

# Calculate the average pack size for both groups
avg_pack_size_target = target_segment_data['PACK_SIZE'].mean()
avg_pack_size_rest = rest_of_population_data['PACK_SIZE'].mean()

# Plot the distribution of pack sizes for both groups
plt.figure(figsize=(10, 6))
sns.kdeplot(target_segment_data['PACK_SIZE'], fill=True, color='blue', label='Target Segment', alpha=0.5)
sns.kdeplot(rest_of_population_data['PACK_SIZE'], fill=True, color='red', label='Rest of Population', alpha=0.5)

plt.title('Pack Size Distribution Comparison')
plt.xlabel('Pack Size')
plt.ylabel('Density')
plt.legend()
plt.show()

# Conduct a t-test to see if there's a significant difference in pack sizes
t_stat, p_value = stats.ttest_ind(target_segment_data['PACK_SIZE'].dropna(),
                                   rest_of_population_data['PACK_SIZE'].dropna())

# Print average pack sizes and t-test result
print(f"Average Pack Size for Target Segment: {avg_pack_size_target}")
print(f"Average Pack Size for Rest of Population: {avg_pack_size_rest}")
print(f"T-test result: t-statistic = {t_stat}, p-value = {p_value}")

# Interpretation
if p_value < 0.05:
    print("There is a statistically significant difference in pack sizes between the target segment and the rest of the population.")
else:
    print("There is no statistically significant difference in pack sizes between the target segment and the rest of the population.")
```

Insights: Our target segment (Mainstream young singles/couples) tend to purchase slightly larger packs of chips compared to the rest of the population.
This insight can inform marketing strategies, such as promoting larger pack sizes or giving tailored offers to this segment.
